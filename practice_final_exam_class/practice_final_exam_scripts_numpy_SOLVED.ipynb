## Exercise 1: Sampling from Distributions (~10 min)

Set a random seed of `42`, then generate:

1. **Exam scores**: 500 samples from a normal distribution with mean=70 and std=10
2. **Random prices**: 500 samples from a uniform distribution between 5.0 and 50.0
3. **Dice rolls**: 500 samples of random integers between 1 and 6 (inclusive)

For each array, print the mean, standard deviation, min, and max.

````python
np.random.seed(42)

# Exam scores — normal distribution
exam_scores = np.random.normal(70, 10, 500)

# Random prices — uniform distribution
prices = np.random.uniform(5.0, 50.0, 500)

# Dice rolls — random integers (1 to 6 inclusive → high=7)
dice_rolls = np.random.randint(1, 7, 500)

for name, arr in [
    ("Exam scores", exam_scores),
    ("Prices", prices),
    ("Dice rolls", dice_rolls),
]:
    print(f"{name}:")
    print(f"  Mean: {arr.mean():.4f}")
    print(f"  Std:  {arr.std():.4f}")
    print(f"  Min:  {arr.min():.4f}")
    print(f"  Max:  {arr.max():.4f}")
    print()
````

Create a file `scripts/functions.py` in your editor/IDE with:

1. A function `min_max_scale(arr)` that applies min-max scaling to a NumPy array:
   - Formula: `(arr - arr.min()) / (arr.max() - arr.min())`
   - Returns a new array with values in [0, 1]

2. An `if __name__ == "__main__":` guard that:
   - Creates a sample array: `np.array([10, 20, 30, 40, 50])`
   - Prints the original array and the scaled result

Then, in this notebook:

- Apply it to the array `np.array([4, 8, 15, 16, 23, 42])`
- Verify that the min of the result is 0 and the max is 1

`scripts/functions.py`:

```python
import numpy as np


def min_max_scale(arr):
    """Apply min-max scaling to a NumPy array. Returns values in [0, 1]."""
    return (arr - arr.min()) / (arr.max() - arr.min())


if __name__ == "__main__":
    sample = np.array([10, 20, 30, 40, 50])
    scaled = min_max_scale(sample)
    print(f"Original: {sample}")
    print(f"Scaled:   {scaled}")
```

## Exercise 4: Distribution Sampler Script (~15 min)

Create a file `scripts/sampler.py` in your editor/IDE that:

1. Takes 3 command-line arguments: `distribution`, `n_samples`, `seed`
2. Supports three distributions:
   - `"normal"` → `np.random.normal(0, 1, n_samples)`
   - `"uniform"` → `np.random.uniform(0, 1, n_samples)`
   - `"randint"` → `np.random.randint(1, 100, n_samples)`
3. Sets the random seed before generating samples
4. Prints:
   - The first 5 values
   - The mean and standard deviation
5. If the wrong number of arguments is provided, prints a usage message and exits
6. If an unknown distribution is given, prints an error message

**Expected usage:**

```bash
python scripts/sampler.py normal 1000 42
```

**Expected output:**
```
Distribution: normal
Samples: 1000 (seed=42)
First 5 values: [ 0.49671415 -0.1382643   0.64768854  1.52302986 -0.23415337]
Mean: 0.0198
Std:  0.9842
```

`scripts/sampler.py`:

```python
import sys
import numpy as np


if __name__ == "__main__":
    if len(sys.argv) != 4:
        print("Usage: python sampler.py <distribution> <n_samples> <seed>")
        print("Distributions: normal, uniform, randint")
        sys.exit(1)

    distribution = sys.argv[1]
    n_samples = int(sys.argv[2])
    seed = int(sys.argv[3])

    np.random.seed(seed)

    if distribution == "normal":
        samples = np.random.normal(0, 1, n_samples)
    elif distribution == "uniform":
        samples = np.random.uniform(0, 1, n_samples)
    elif distribution == "randint":
        samples = np.random.randint(1, 100, n_samples)
    else:
        print(f"Error: Unknown distribution '{distribution}'")
        print("Supported: normal, uniform, randint")
        sys.exit(1)

    print(f"Distribution: {distribution}")
    print(f"Samples: {n_samples} (seed={seed})")
    print(f"First 5 values: {samples[:5]}")
    print(f"Mean: {samples.mean():.4f}")
    print(f"Std:  {samples.std():.4f}")
```

## Exercise 5: Synthetic Data Generator Script (~15 min)

Create a file `scripts/generate_data.py` in your editor/IDE that:

1. Contains a function `generate_dataset(n_rows, seed)` that:
   - Sets the random seed
   - Creates a DataFrame with these columns:
     - `date`: daily dates starting from `"2025-01-01"`
     - `temperature`: normal distribution (mean=22, std=4)
     - `humidity`: uniform distribution (30, 90)
     - `visitors`: random integers (50, 500)
   - Returns the DataFrame

2. Has an `if __name__ == "__main__":` guard that:
   - Takes 2 command-line arguments: `n_rows` and `seed`
   - Calls `generate_dataset(n_rows, seed)`
   - Saves the result to `output.csv`
   - Prints the shape and summary statistics
   - Shows a usage message if wrong number of arguments

**Expected usage:**

```bash
python scripts/generate_data.py 100 42
```

Then, in this notebook:
- Import `generate_dataset` from `generate_data`
- Call it with `n_rows=50, seed=42`
- Print the first 5 rows

`scripts/generate_data.py`:

```python
import sys
import numpy as np
import pandas as pd


def generate_dataset(n_rows, seed):
    """Generate a synthetic dataset with weather and visitor data."""
    np.random.seed(seed)
    df = pd.DataFrame({
        "date": pd.date_range("2025-01-01", periods=n_rows, freq="D"),
        "temperature": np.random.normal(22, 4, n_rows),
        "humidity": np.random.uniform(30, 90, n_rows),
        "visitors": np.random.randint(50, 500, n_rows),
    })
    return df


if __name__ == "__main__":
    if len(sys.argv) != 3:
        print("Usage: python generate_data.py <n_rows> <seed>")
        print("Example: python generate_data.py 100 42")
        sys.exit(1)

    n_rows = int(sys.argv[1])
    seed = int(sys.argv[2])

    df = generate_dataset(n_rows, seed)
    df.to_csv("output.csv", index=False)

    print(f"Generated dataset with shape: {df.shape}")
    print(f"Saved to output.csv")
    print()
    print(df.describe())
```
