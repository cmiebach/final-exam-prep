## Solution 9

```python
class Pokemon:
    def __init__(self, name, type_name, hp):
        self.name = name
        self.type_name = type_name
        self.hp = hp

    def show_info(self):
        print(f"Name: {self.name}")
        print(f"Type: {self.type_name}")
        print(f"HP: {self.hp}")

    def take_damage(self, amount):
        self.hp -= amount

pikachu = Pokemon("Pikachu", "electric", 100)
pikachu.show_info()
pikachu.take_damage(15)
pikachu.show_info()
```

## Solution 13

```python
def pokemon_summary(name):
    url = f"https://pokeapi.co/api/v2/pokemon/{name.lower()}"
    response = requests.get(url)
    data = response.json()
    pokemon_name = data["name"].capitalize()
    pokemon_type = data["types"][0]["type"]["name"]
    height = data["height"]
    weight = data["weight"]
    return f"{pokemon_name} is an {pokemon_type} Pokémon with height {height} and weight {weight}."

print(pokemon_summary("pikachu"))
```

## Solution 33

```python
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Data Explorer", layout="wide")
st.title("Sales Data Explorer")

df["revenue"] = df["quantity"] * df["unit_price"]
selected_categories = st.sidebar.multiselect("Categories:", options=df["category"].unique().tolist(), default=df["category"].unique().tolist())
selected_regions = st.sidebar.multiselect("Regions:", options=df["region"].unique().tolist(), default=df["region"].unique().tolist())
filtered = df[(df["category"].isin(selected_categories)) & (df["region"].isin(selected_regions))]

col1, col2, col3 = st.columns(3)
col1.metric("Total Revenue", f"${filtered['revenue'].sum():,.2f}")
col2.metric("Number of Orders", f"{len(filtered):,}")
col3.metric("Avg Order Value", f"${filtered['revenue'].mean():,.2f}")

st.dataframe(filtered, use_container_width=True)

revenue_by_category = filtered.groupby("category")["revenue"].sum().reset_index()
fig = px.bar(revenue_by_category, x="category", y="revenue", color="category")
st.plotly_chart(fig, use_container_width=True)
```
