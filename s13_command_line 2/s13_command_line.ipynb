## 1. What is the Command Line?

Up until now, you've been working inside Jupyter notebooks — a graphical, point-and-click environment. But to run Python scripts, install packages, and use many developer tools, you need to use the **command line**.

Don't worry, you only need a handful of commands. The terminal is just another way to talk to your computer — instead of clicking icons, you type short text commands.

### GUI vs CLI

There are two main ways to interact with your computer:

| | GUI (Graphical User Interface) | CLI (Command Line Interface) |
|---|---|---|
| **How you interact** | Click icons, drag files, use menus | Type text commands |
| **Examples** | Finder (Mac), File Explorer (Windows) | Terminal (Mac), Command Prompt (Windows) |
| **Best for** | Browsing, everyday tasks | Running scripts, automation, developer tools |
| **Learning curve** | Very low — intuitive | Slightly higher — you need to know the commands |
| **Power & flexibility** | Limited to what the interface offers | Very flexible — can automate, chain commands |
| **Used by** | Everyone | Developers, data scientists, system admins |

### Why Learn the Command Line?

You might be wondering: *"If I can do everything with a GUI, why bother with the terminal?"*

Here are the main reasons for this course:

- **Running Python scripts**: Running `.py` files from the terminal with `python script.py`
- **Running web apps**: Launching Streamlit apps from the terminal (next session!)
- **Package management**: Installing Python libraries with `pip install`
- **Git version control**: Committing and pushing code with `git`
- **Automation**: Running scripts on a schedule, processing files in bulk
- **It's everywhere**: Cloud servers, job interviews, documentation — the terminal is the lingua franca of software development

### Terminology

You'll hear several names used interchangeably. They all refer to roughly the same thing:

- **Terminal** — the application window where you type commands
- **Command line** — the text-based interface itself
- **Shell** — the program that interprets your commands (e.g., Bash, Zsh, PowerShell)
- **Console** — another word for terminal
- **CLI** — Command Line Interface (the opposite of GUI)

For our purposes, we'll use "terminal" to refer to the application, and "command line" to refer to the interface in general.

## 3. Navigating the File System

When you open the terminal, you're always "inside" a folder (directory) on your computer. Think of it like having a Finder/File Explorer window open — except instead of seeing icons, you type commands to look around and move between folders.

### Essential Navigation Commands

| Action | Mac / Linux | Windows | Notes |
|--------|-------------|---------|-------|
| Where am I? | `pwd` | `cd` | Prints your current folder |
| List files | `ls` | `dir` | Shows files and folders here |
| Go into a folder | `cd folder_name` | `cd folder_name` | Same on both! |
| Go back one level | `cd ..` | `cd ..` | Same on both! |
| Go to home | `cd ~` | `cd %USERPROFILE%` | Jumps to your home folder |
| Go to a full path | `cd /full/path` | `cd C:\full\path` | Mac uses `/`, Windows uses `\` |

> **Good news:** Most commands (`cd`, `mkdir`, `python`) work **identically** on Mac and Windows. The main differences are `ls` vs `dir`, `pwd` vs `cd`, and `/` vs `\` in paths.

## 5. Running Python from the Terminal

You've been running Python code inside Jupyter notebooks, where you execute one cell at a time. But Python can also run `.py` files directly from the terminal.

### Running a Script

```
python script.py
```

> **Mac note:** On some Macs, you may need `python3` instead of `python`. Try both — whichever works is fine.

### How It Differs from Jupyter

| | Jupyter Notebook | Terminal (`python script.py`) |
|---|---|---|
| **Execution** | One cell at a time, in any order | Entire file, top to bottom |
| **Output** | Shows below each cell | Prints to the terminal |
| **State** | Variables persist between cells | Variables exist only while the script runs |
| **Interactivity** | Interactive — edit and re-run cells | Runs once and exits |
| **File type** | `.ipynb` | `.py` |

The key difference: when you run `python script.py`, the entire file executes from the first line to the last line, then the program ends. There's no "run this cell" — it's all or nothing.

### Try It: Run the Practice Scripts

From the `practice/` folder, navigate to `scripts/` and run them. These commands are the same on Mac and Windows:

| Step | Command | Expected output |
|------|---------|-----------------|
| Go to scripts folder | `cd scripts` | |
| Run the hello script | `python hello.py` | `Hello from the terminal!` ... |
| Run the interactive script | `python greet.py` | Asks for your name, then greets you |
| Go back | `cd ..` | Back in `practice/` |

> If `python hello.py` gives an error on Mac, try `python3 hello.py`.

Notice how `greet.py` uses `input()` to ask for your name — this is how terminal programs get user input, instead of widgets.

### Checking Your Python Version

```
python --version
```

You should see something like `Python 3.11.5`. On Mac, you may need `python3 --version`.

### Installing Packages

You've probably used `pip install` before, perhaps from inside a notebook cell with `!pip install`. From the terminal, it's the same command without the `!`:

```
pip install pandas
pip install streamlit
pip install plotly
```

Or if you're using `uv` (our recommended tool):

```
uv add pandas
uv add streamlit
```

These commands work the same on Mac and Windows.

## 7. Common Mistakes and Tips

The terminal can be confusing at first. Here are the most common errors and how to fix them.

### Common Errors

| Error Message | Cause | Fix |
|---------------|-------|-----|
| `No such file or directory` | You're in the wrong folder, or the file name is misspelled | Use `pwd` (Mac) or `cd` (Win) to check where you are, then `cd` to the correct folder |
| `python: command not found` | Python is not installed or not in PATH | Install Python, or try `python3` instead of `python` |
| `ModuleNotFoundError` | A package is not installed | Run `pip install package_name` |
| `Permission denied` | Mac/Linux file permissions issue | Try `sudo pip install ...`, or check file permissions |
| `SyntaxError` when running a script | Typo or bug in your `.py` file | Open the file in VS Code and check the line number shown in the error |
| Terminal seems stuck / won't accept input | A program is still running | Press `Ctrl+C` to stop it |

### Pro Tips

**Tab completion** — This is the most useful trick in the terminal. Start typing a file or folder name and press `Tab`:

```bash
# In the practice folder:
cd rep<Tab>
# Autocompletes to: cd reports/

cat reports/q1<Tab>
# Autocompletes to: cat reports/q1_sales.txt

cd scr<Tab>
# Autocompletes to: cd scripts/
```

This saves typing and helps avoid typos. If there are multiple matches, press `Tab` twice to see all options:

```bash
cat reports/q<Tab><Tab>
# Shows: q1_sales.txt  q2_sales.txt  q3_sales.txt
```

**Command history** — Press the `Up Arrow` key to cycle through previously typed commands. This is great for re-running a script without retyping the command.

**Copy and paste:**

| Mac | Windows (Command Prompt) | Windows (PowerShell) |
|-----|--------------------------|----------------------|
| `Cmd+V` | Right-click | `Ctrl+V` |

**Spaces in folder names** — If a folder name has spaces, wrap it in quotes:

```bash
cd "My Folder Name"
```
